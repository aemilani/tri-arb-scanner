```python
import requests
```

```python
STABLECOINS = {'USDT', 'USDC', 'FDUSD', 'TUSD', 'USDP', 'DAI', 'BUSD', 'AEUR'}
```

```python
def is_stablecoin(coin_ticker):
    return coin_ticker.upper() in STABLECOINS
```

```python
url = 'https://api.binance.com/api/v3/exchangeInfo'
data = requests.get(url).json()
symbols = data['symbols']
spot_pairs = [symbol['symbol'] for symbol in symbols if symbol['isSpotTradingAllowed']]
```

```python
arb_pairs = dict()
for stable in STABLECOINS:
    arb_pairs[stable] = []
    coins = [symbol.replace(stable, '') for symbol in spot_pairs if stable in symbol]
    for coin1 in coins:
        for coin2 in coins:
            pair = coin1+coin2
            if pair in spot_pairs:
                arb_pairs[stable].append(pair)
```

```python
for stable in STABLECOINS:
    print(f"{stable}: {len(arb_pairs[stable])}")
```
